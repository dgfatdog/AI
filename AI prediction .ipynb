```python
#Import tensorflow load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.losses import sparse_categorical_crossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential, save_model, load_model
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
import numpy as np
import os
```

```python
#Define a function to load the model

def get_model():
    model_loaded = False
    while not model_loaded:
        model_path = input("Please enter the file path of the pretrained tensorflow model: ")
        try:
            model = load_model(model_path, compile = True)
            model_loaded = True
        except:
            print("Model loading failed. Please try again.")

    return model
```

```python
#Define a function a locate directory

def get_img_dir():
    image_loaded = False
    while not image_loaded:
        img_dir_path = input("Please enter the image directory path of the images you would like to classify: ")
        try:
            for img_path in os.listdir(img_dir_path):
                Image.open(os.path.join(img_dir_path, img_path))
            image_loaded = True
        except:
            print("Not all files in the directory are images, or some images cannot be opened.")
            clear_bad_files(img_dir_path)
    return img_dir_path
```

```python
def clear_bad_files(img_dir):

    bad = 0   
    good = 0  
    moved = 0

    for root, dirs, files in os.walk(img_dir, topdown=False):
        for name in files:
            try:
                img = Image.open(os.path.join(root, name)) #try to open the image file
                img.verify() 
                print('Good file: ' + os.path.join(root, name))
                good = good + 1
            except Exception as e:
                bad = bad + 1
                print('Bad file: ' + os.path.join(root, name)) #print out the filenames of corrupted files
                if not os.path.isdir(os.path.join(root, "00 Bad Files")):
                    os.makedirs(os.path.join(root, "00 Bad Files"))
                    os.rename(os.path.join(root, name), os.path.join(os.path.join(root, "00 Bad Files"), name))
                else:
                    os.rename(os.path.join(root, name), os.path.join(os.path.join(root, "00 Bad Files"), name))
                print('Moved: '+ os.path.join(root, name))
                moved = moved + 1
        
        print(f"Number of bad files: {bad}")
        print(f"Number of good files: {good}")
        print(f"Number of files moved: {moved} ")
```

```python
def get_dimensions():
    dim_1 = int(input("Please input the first input dimension of the model: "))
    dim_2 = int(input("Please input the second input dimension of the model: "))
    return dim_1, dim_2
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
img_dir = get_img_dir()
model = get_model()
dim_1, dim_2 = get_dimensions()

samples_to_predict = []

for img_path in os.listdir(img_dir):
    im = keras.preprocessing.image.load_img(os.path.join(img_dir, img_path), target_size=(dim_1, dim_2)) 
    img_array = keras.preprocessing.image.img_to_array(im)
    img_array = tf.expand_dims(img_array, 0)
    Image.open(os.path.join(img_dir, img_path))
    prediction = model.predict(img_array)

    class_name = prediction['category'][0].decode()
    # print(type(img_dir))
    # print(type(prediction['category'][0]))
    print(os.path.join(img_dir, class_name))
    if not os.path.isdir(os.path.join(img_dir, class_name)):
        os.makedirs(os.path.join(img_dir, class_name))
        os.rename(os.path.join(img_dir, img_path), os.path.join(os.path.join(img_dir, class_name), img_path))
    else:
        os.rename(os.path.join(img_dir, img_path), os.path.join(os.path.join(img_dir, class_name), img_path))
    
    # # print(prediction)
    # # print(type(prediction))
    # # print(prediction['category'])
    # # print(type(prediction['category']))
    print(prediction['category'][0])
```
